# Matplotlibの学習
2次元のグラフを描写するためのライブラリー。データを可視化する段階で使用する。 今回はグラフを描写するコードオブジェクト指向インターフェースを用いる

```python
#インポート
import matplotlib.pyplot as plt
import matplotlib.style

#ggplotstyleを指定する
matplotlib.style.use("ggplot")

x = [1,2,3]
y = [2,4,9]

plt.plot(x,y)
plt.title("pyplot")

plt.show()
```

## 描写オブジェクトを指定

```python
#描写オブジェクト - 2つのサブプロットを配置する
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2)
plt.show()
```

```python
fig, axes = plt.subplots(ncols=2)
plt.show()
```

```python
fig, axes = plt.subplots(2,2)
plt.show()
```

```python
#タイトル、軸ラベル
fig, axes = plt.subplots(ncols=2)

#サブプロットのタイトル
axes[0].set_title("Subplot Title 01")
axes[1].set_title("Subplot Title 02")

#描写オブジェクトにタイトルを設定
fig.suptitle("Figure Title")
plt.show()
```

```python
#凡例 - サブプロットに凡例を表示する
fig, ax = plt.subplots()

ax.plot([1,2,3],[2,4,9], label="Legend Label")
ax.legend(loc="best")

plt.show()
```

```python
# 折れ線グラフを描く
fig, ax = plt.subplots()

x = [1,2,3]
y1 = [1,2,3]
y2 = [3,1,2]

ax.plot(x,y1)
ax.plot(x,y2)

plt.show()
```

```python
# sin,cosカーブを描く
import numpy as np

x = np.arange(0.0,15.0,0.1)
y1 = np.sin(x)
y2 = np.cos(x)

fig,ax = plt.subplots()
ax.plot(x,y1,label="sin")
ax.plot(x,y2,label="cos")
ax.legend()

plt.show()
```

## 棒グラフ

```python
#棒グラフ
fig,ax = plt.subplots()

x = [1,2,3]
y = [10,2,3]

#棒グラフの描写
ax.bar(x,y)
plt.show()
```

```python
#棒グラフ
fig,ax = plt.subplots()

x = [1,2,3]
y = [10,2,3]

#棒グラフの描写 - ラベルを指定し、barの引数にする
labels = ["spam","ham","egg"]
ax.bar(x,y, tick_label=labels)

plt.show()
```

```python
#棒グラフ
fig,ax = plt.subplots()

x = [1,2,3]
y = [10,2,3]

#棒グラフの描写 - 横方向のバーにする(barh)
labels = ["spam","ham","egg"]
ax.barh(x,y, tick_label=labels)

plt.show()
```

## 複合棒グラフ

```python
#棒グラフ
fig,ax = plt.subplots()

x = [1,2,3]
y1 = [10,2,3]
y2 = [5,3,6]

#y1・y2を同じコラムに描く複合棒グラフにしたい場合、幅分をすらして棒グラフを描写する
#棒グラフの描写 - 横方向のバーにする(barh)
width=0.4

labels = ["spam","ham","egg"]
ax.bar(x,y1, width=width, tick_label=labels, label="y1")

#y2のx方向の位置をwidthの分ずらす
x2=[num+width for num in x]
ax.bar(x2,y2, width=width, tick_label=labels, label="y2")

plt.show()
```
